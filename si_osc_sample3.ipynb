```
import numpy as np
import matplotlib.pyplot as plt
```

```
sam_rate = 44000 #Hz
frequency = 1000 #Hz
```

```
f = np.arange(1,sam_rate,1)
```

```
tone = sin(2*frequency*(f/sam_rate))
```

```
plt.xlim(1,1000)
plt.plot(tone)
```
